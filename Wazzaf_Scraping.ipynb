```python
# first page
driver = webdriver.Chrome()
driver.get(url)

job_titles = driver.find_elements(By.XPATH, "//h2/a")
company_name = driver.find_elements(By.XPATH, "//div[@class='css-d7j1kk']/a")
company_location = driver.find_elements(By.XPATH, "//div[@class='css-d7j1kk']/span")
the_job_types = driver.find_elements(By.XPATH, "//span[@class = 'css-1ve4b75 eoyjyou0']")
post = driver.find_elements(By.XPATH, "//div[@class='css-4c4ojb' or @class='css-do6t5g']")
skills = driver.find_elements(By.XPATH, "//div[@class = 'css-y4udm8']/div/a[@class='css-o171kl' or @class='css-5x9pm1']")

# inner page
```

```python
# Get All Links in each page

driver = webdriver.Chrome()
driver.get(url)

link = driver.find_elements(By.XPATH, "//h2/a")

for i in link:
    links.append(i.get_attribute('href'))
```

```python
# For Salary

import time
driver = webdriver.Chrome()

for url in links:
    driver.get(url)
    time.sleep(3)  # Wait for the page to load, adjust the time if necessary

    div_elements_Salary = driver.find_elements(By.CLASS_NAME, "css-rcl8e5")
    
    
    for outer_div in div_elements_Salary:
        first_span_Salary = outer_div.find_element(By.CLASS_NAME, "css-wn0avc")
        
        if first_span_Salary.text.strip() != 'Salary:':
            continue

        outer_span_Salary = outer_div.find_element(By.CLASS_NAME, "css-47jx3m")
        
        inner_span_Salary = outer_span_Salary.find_element(By.CLASS_NAME, "css-4xky9y")
        
        Salary_value = inner_span_Salary.text.strip()
        print(Salary_value)
        
        
```

# Scrap From Multiple Pages

```python

while True:
    url = f'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=Data%20Analyst&start={page_num}'
    driver = webdriver.Chrome()
    driver.get(url)
    
    page_limit = int(driver.find_element(By.XPATH, "//span[@class='css-xkh9ud']/strong").text)
    try:
        if (page_num > page_limit // 15):
            print('The end')
            break
        
        
        job_titles = driver.find_elements(By.XPATH, "//h2/a")
        company_name = driver.find_elements(By.XPATH, "//div[@class='css-d7j1kk']/a")
        company_location = driver.find_elements(By.XPATH, "//div[@class='css-d7j1kk']/span")
        the_job_types = driver.find_elements(By.XPATH, "//span[@class = 'css-1ve4b75 eoyjyou0']")
        post = driver.find_elements(By.XPATH, "//div[@class='css-4c4ojb' or @class='css-do6t5g']")
        skills = driver.find_elements(By.XPATH, "//div[@class = 'css-y4udm8']/div/a[@class='css-o171kl' or @class='css-5x9pm1']")
        
        
        for i in job_titles:
            links.append(i.get_attribute('href'))
        
        for i in range(len(job_titles)):
            titles.append(job_titles[i].text)
            companies.append(company_name[i].text)
            locations.append(company_location[i].text)
            job_types.append(the_job_types[i].text)
            posteds.append(post[i].text)
            job_skills.append(skills[i].text)
            
        page_num += 1
        print('page Switched')
    except:
        print('u got an error')
        
    driver.quit()
        
```

```python
import time
driver = webdriver.Chrome()


for urls in links:
    
    try:
        driver.get(urls)
        time.sleep(3)  # Wait for the page to load, adjust the time if necessary
        
        ########################## For Requirements #########################
        
        requirements_sections = driver.find_elements(By.CLASS_NAME, "css-ghicub")
        for section in requirements_sections:
            title = section.find_element(By.CLASS_NAME, "css-fwj1k5")

            if title.text != 'Job Requirements':
                continue
            
            div = section.find_element(By.CLASS_NAME, "css-1t5f0fr")
        
            uls_fo_div = div.find_elements(By.TAG_NAME, "ul")
            for ul in uls_fo_div:
                lis = ul.find_elements(By.TAG_NAME, "li")
                for li in lis:
                    job_requirements.append(li.text)
                    
        ########################## For Requirements #########################
        
        ########################## For describtions #########################
        
        Discrption_sections = driver.find_elements(By.CLASS_NAME, "css-ghicub")
        for section in Discrption_sections:
            title = section.find_element(By.CLASS_NAME, "css-fwj1k5")

            if title.text != 'Job Description':
                continue
            
            div = section.find_element(By.CLASS_NAME, "css-1uobp1k")
        
            uls_fo_div = div.find_elements(By.TAG_NAME, "ul")
            for ul in uls_fo_div:
                lis = ul.find_elements(By.TAG_NAME, "li")
                for li in lis:
                    job_descriptions.append(li.text)
                    
        ########################## For describtions #########################
        
        ######################## For Salary #########################
        
        div_elements_Salary = driver.find_elements(By.CLASS_NAME, "css-rcl8e5")
        for outer_div in div_elements_Salary:
            first_span_Salary = outer_div.find_element(By.CLASS_NAME, "css-wn0avc")
            
            if first_span_Salary.text.strip() != 'Salary:':
                continue

            outer_span_Salary = outer_div.find_element(By.CLASS_NAME, "css-47jx3m")
            
            inner_span_Salary = outer_span_Salary.find_element(By.CLASS_NAME, "css-4xky9y")
            
            Salary_value = inner_span_Salary.text.strip()
            job_Salary.append(Salary_value)
            
        ####################### For Salary #########################
        
        print('link processed')
    except:
        print('YOU HAVE AN ERROR')
    
            
```

# put the data into csv file

```python
with open('E:\Data_Jobs.csv', 'w', newline='') as file:
    wr = csv.writer(file)
    wr.writerow(['Company Name', 'Job Title', 'Job Type', 'Location', 'Posted', 'Job Skills', 'Salary', 'requirements', 'describtion'])
    wr.writerows(zip(companies, titles, job_types, locations, posteds, job_skills, job_Salary, job_requirements, job_descriptions))
```
